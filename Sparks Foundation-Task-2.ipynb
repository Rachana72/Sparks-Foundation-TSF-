###  Finding the optimal number of clusters for K-Means and determining the value of K and Plotting the graph onto a line graph to observe the pattern.

```python
from sklearn.cluster import KMeans

within_class = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    #print(kmeans.inertia_)
    within_class.append(kmeans.inertia_)
    
```

```python
# Plotting the results onto a line graph, 
# `allowing us to observe 'The elbow'
plt.plot(range(1, 11), within_class)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Within class sum of square')
plt.show()
```

### Creating K-Means Classifier by predicting 4 features

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 50, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 50, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 50, c = 'green', label = 'Iris-virginica')
plt.legend()
```
